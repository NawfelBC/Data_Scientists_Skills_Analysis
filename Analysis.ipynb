```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud 
import numpy as np
from PIL import Image
import random
```

```python
#Load the data
# from google.colab import files
# uploaded = files.upload()
df = pd.read_csv('skills.csv')
df
```

```python
#Plot the data into histogram
df['Skills'].value_counts()[:20].plot(kind='barh')
```

```python
#Show WordCloud
def color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(10, 20)
text = df.Skills.str.cat(sep=' ')
stopwords = ['de','des']
wordcloud = WordCloud(background_color='white', width=2000 , height=1000, max_words=75, stopwords=stopwords).generate(text) 
plt.figure(figsize=(20,20))
plt.imshow(wordcloud.recolor(color_func=color_func, random_state=3),
           interpolation="bilinear")
plt.axis('off')
plt.show()
wordcloud.to_file("wordcloud.png")
```
